#Splitting data

```python
from datasets import load_dataset


# Load the entire dataset from the specified file in Google Drive
full_dataset = load_dataset('text', data_files={'data': f'{base_dir}/text.txt'})

# Determine the size for the training subset (80% of the full dataset)
train_size = int(0.8 * len(full_dataset['data']))

# Splitting the full dataset into training and validation subsets
train_subset = full_dataset['data'].select(list(range(train_size)))
val_subset = full_dataset['data'].select(list(range(train_size, len(full_dataset['data']))))

# Specifying paths where the split datasets (training and validation) will be saved
train_path = f"{base_dir}/train.txt"
val_path = f"{base_dir}/validation.txt"

# Save the training subset to its designated file
with open(train_path, 'w') as train_file:
    for item in train_subset:
        train_file.write(item['text'] + '\n')

# Save the validation subset to its designated file
with open(val_path, 'w') as val_file:
    for item in val_subset:
        val_file.write(item['text'] + '\n')
```
